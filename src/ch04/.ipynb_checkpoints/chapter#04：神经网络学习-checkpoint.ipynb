# 从数据中学习

神经网络的一个特征是：可以从数据中进行学习。

神经网络的"从数据中学习"是指通过训练数据来自动调整神经网络的权重和偏置，以使其能够对输入数据进行准确的预测或分类。这种学习过程称为训练，它基于输入数据和相应的目标输出数据之间的关系来调整神经网络的参数。

在训练过程中，神经网络通过反向传播算法计算损失函数的梯度，并使用优化算法（如梯度下降）来最小化损失函数。通过迭代训练样本，神经网络逐渐调整权重和偏置，以提高其对新数据的泛化能力。

通过从大量的训练数据中学习，神经网络可以捕捉到输入数据中的模式和特征，并将其用于对未知数据的预测。这种能力使得神经网络在各种任务中表现出色，如图像识别、语音识别、自然语言处理等。

总而言之，神经网络的"从数据中学习"是指通过训练数据来自动调整网络参数，以使其能够对新数据进行准确的预测或分类。这种学习过程使得神经网络能够从数据中提取有用的信息，并应用于实际应用中。

在计算机视觉中：常用的特征量包括SIFT，SURF和HOG等，使用这些特征量将图像数据转换为向量，然后对转换后的向量使用机器学习中的SVM，KNN等分类器进行学习。

数据可以被分为两个部分：

# 损失函数

神经网络的学习中所使用的指标成为 损失函数（loss function），这个损失函数可以使用任意函数，但是一般使用均方误差和交叉熵误差等。

损失函数是表示神经网络性能的“恶劣程度”的指标，即当前的神经网络对监督数学在多大程度上是不拟合的，在多大程度上不一致。

这个值越小，也就意味着损失越少，模型的表现效果越好。
