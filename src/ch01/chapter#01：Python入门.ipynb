```python
# 类
# 类的定义
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def say_hello(self):
        print('Hello, my name is %s, I am %d years old' % (self.name, self.age))
# 类的实例化
p = Person('Tom', 20)
p.say_hello()
```

```python
# 类的继承
class Student(Person):
    def __init__(self, name, age, grade):
        super(Student, self).__init__(name, age)
        self.grade = grade
    def say_hello(self):
        print('Hello, my name is %s, I am %d years old, I am in grade %d' % (self.name, self.age, self.grade))
s = Student('Jerry', 18, 3)
s.say_hello()
```

```python
# 子类无法访问父类的私有变量
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
        
    def __say_hello(self):
        print('Hello, my name is %s, I am %d years old' % (self.name, self.age))
        
    def say_hello(self):
        self.__say_hello()
        
class Student(Person):
    def __init__(self, name, age, grade):
        super(Student, self).__init__(name, age)
        self.grade = grade
    def say_hello(self):
        print('Hello, my name is %s, I am %d years old, I am in grade %d' % (self.name, self.age, self.grade))
        
# 展示关于私有变量的问题
p = Person('Tom', 20)
# 顺便捕捉一下一场
try:
    p.__say_hello()
except AttributeError as e:
    print(e)
```

```python
# 异常处理
try:
    x = 1 / 0
except ZeroDivisionError as e:
    print(e)
```

```python
# 矩阵乘法
x = np.array([[1, 2, 3], [4, 5, 6]])
y = np.array([[1, 2], [3, 4], [5, 6]])
print(np.dot(x, y))
```

矩阵乘法：$C_{ij} = \sum_{k=1}^{n} A_{ik}B_{kj}$

解释：$C_{ij}$是矩阵$C$的第$i$行第$j$列的元素，$A_{ik}$是矩阵$A$的第$i$行第$k$列的元素，$B_{kj}$是矩阵$B$的第$k$行第$j$列的元素，$n$是矩阵$A$的列数，也是矩阵$B$的行数。

矩阵乘法的前提是矩阵$A$的列数等于矩阵$B$的行数。

矩阵乘法的几何意义：矩阵$A$可以看作是一个线性变换，矩阵$B$可以看作是一个坐标，矩阵$C$是变换后的坐标。

```python
# 绘制三维图
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x ** 2 + y ** 2))
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z)
plt.show()
```

```python
# 绘制三维散点图
x = np.random.random(100)
y = np.random.random(100)
z = np.random.random(100)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z)
plt.show()
```

```python
# 绘制三维曲线
x = np.linspace(-5, 5, 100)
y = np.sin(x)
z = np.cos(x)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, y, z)
plt.show()
```
